```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(613)

# Generate an array of 50 sequential x-values (not used in the histogram)
x = np.arange(50)

# Generate 50 random integer y-values between 0 and 100
y = np.random.randint(0, 100, 50)

# Create a figure and axes object with a specific size (width=5, height=3 inches)
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a histogram of the y-values (distribution of values from 0 to 100)
ax.hist(y)

# Optional: if you'd like to see the scatter plot instead, uncomment the next line
# ax.scatter(x, y)

# Display the plot
plt.show()
```

```python
# Create a figure and a single subplot with specified figure size (width=5, height=3 inches)
fig, ax = plt.subplots(figsize=(5, 3))

# Plot y versus x as a line chart (showing growth over time)
ax.plot(x, y)

# Set the title of the plot
ax.set_title('Total growth over time')

# Label the y-axis to indicate what is being measured
ax.set_ylabel('Total growth')

# Label the x-axis to indicate time progression
ax.set_xlabel('Years since start')
```

```python
# Create a figure and a single subplot with specified size (5 inches wide by 3 inches tall)
fig, ax = plt.subplots(figsize=(5, 3))

# Create a scatter plot of y versus x using star-shaped markers in indigo color
ax.scatter(x, y, marker='*', color='indigo')
```

```python
# Create a figure and axis with a custom size (5 inches wide by 3 inches tall)
fig, ax = plt.subplots(figsize=(5, 3))

# Plot a line graph with star markers, indigo color, dashed lines, and line width of 2
ax.plot(x, y, marker='*', color='indigo',
        linestyle='--', linewidth=2)
```
